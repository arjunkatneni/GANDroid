```python

from numpy import loadtxt
from numpy import unique
import pandas as pd
from sklearn.preprocessing import LabelEncoder
# load the dataset
benign_set = pd.read_csv("../input/ml-final-dataset/PlayStore Application.csv")
malicious_set = pd.read_csv("../input/ml-final-dataset/Malware_Data.csv")
benign_set['label'] = 0
malicious_set['label'] = 1
frames = [benign_set, malicious_set]
combined_set = pd.concat(frames)
print(combined_set.shape)
uniq_count = combined_set.nunique()
to_del = [i for i,v in enumerate(uniq_count) if v == 1]
print(to_del)
combined_set.drop(combined_set.columns[to_del], axis=1, inplace=True)
print(combined_set.shape)
le = LabelEncoder()
combined_set['Category'] = le.fit_transform(combined_set['Category'].astype(str))
combined_set.drop(['Package'], axis=1, inplace=True)
print(combined_set.isnull().sum())
#combined_set = combined_set.fillna(0)
combined_set.head()
```
